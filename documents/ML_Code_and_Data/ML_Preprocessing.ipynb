## Combine the 1999 and the 2011 datasets into one 

```python
# stack one df on top of another; frist_df then second_df
combined_df = pd.concat([first_df, second_df])
combined_df['CHILDID'] = range(1, len(combined_df) + 1)
combined_df['PARENTID'] = range(1, len(combined_df) + 1)
combined_df.head()
```

```python
pd.set_option('display.max_rows', None) # shows all the columns/rows instead of truncated list 
```

```python
#NOTE: to detect the blank values too, we need to change the objects to numeric (only for the columns we need)
col_names = combined_df.columns.tolist()
col_names.remove('CHILDID')
col_names.remove('PARENTID')
col_names.remove('DOBYY')
col_names.remove('GENDER')
col_names.remove('RACE')
col_names #currently has the names of the columns we want to make numeric 

#convert the desired columns to numeric 
for x in col_names:
    combined_df[x] = pd.to_numeric(combined_df[x], errors='coerce') 
```

Remainder: our target columns are ACTIVE, ATTENTI, BEHAVE, SAMEAG, and SOLVE (we'll be alternating between target labels when we begin testing)
- ACTIVE  =  CHD AS ACTIVE AS SAME-AGE CHILDREN
- ATTENTI =  AS ATTENTIVE AS SAME-AGE CHILDREN
- BEHAVE  =  BEHAVES AS WELL AS OTHER CHILDREN
- SAMEAG  =  AS INDEPENDENT AS SAME-AGE CHILDREN
- SOLVE   =  AS CLEVER AS SAME-AGE CHILDREN

```python
#drop rows with 59 or more missing values
combined_df.dropna(thresh=46, axis=0, inplace=True) 
threshold = int(0.6 * len(combined_df))
combined_df = combined_df.dropna(thresh=threshold, axis=1)

#count the number of datapoints we have
print(f"Length: {len(combined_df)}") 
```

Given that the values within the features are actually categorically we need to...
- convert the columns from numeric to categorical
- find the mode, to fill in the empty values

```python
for col in combined_df.columns[5:]:
    # Check if the column is numeric or has categorical values
    if not combined_df[col].empty:
        mode_value = combined_df[col].mode()
        
        # Check if mode_value is not empty before accessing [0]
        if not mode_value.empty:
            mode_value = mode_value[0]  # Get the first mode value
            
            # Drop the column if the mode is negative
            if mode_value < 0:
                combined_df.drop(col, axis=1, inplace=True)
            else:
                # Fill missing values with the mode
                combined_df[col] = combined_df[col].fillna(mode_value)
        else:
            print(f"Column '{col}' has no mode (all values are missing).")
    else:
        print(f"Column '{col}' is empty.")
```

```python
#checking if correctly replaced the empty values with mode

# Drop 'DOBYY' if it's now a column
print(combined_df.columns.tolist())
combined_df.drop(columns=['CHILDID'], inplace=True, errors='ignore')
combined_df.drop(columns=['PARENTID'], inplace=True, errors='ignore')
combined_df.drop(columns=['DOBYY'], inplace=True, errors='ignore')

combined_df.to_csv('new_data.csv', index=False)
```

# Find the correlation between variables with PEARSON CORRELATION

Pearson Correlation describes the strength in the relationship between variables (ie how are variables related).

the linear relationship shows a r-value between -1 to 1:

- "1" means a positive correlation ("strongest")
- "0" means no correlation
- "-1" means a negative correlation ("weakest")

```python
import numpy as np
import pandas as pd

# Read the data
combined_df = pd.read_csv('new_data.csv')

# Define invalid values to be replaced with NaN
invalid_values = [-9, -8, -7, -1, 0]
#Replace invalid values in numeric columns only

for col in combined_df.columns:
    if np.issubdtype(combined_df[col].dtype, np.number):  # Check if column is numeric
        mode = combined_df[col].mode()[0]
        #combined_df[col] = combined_df[col].replace(invalid_values, np.nan)
        combined_df[col] = combined_df[col].replace(invalid_values, mode)
        #print(f'mode for {col} = {mode}, number empty = {combined_df[col].isna().sum()}')
        print(f"{combined_df[col].value_counts()}\n")


#Convert all columns to categorical type
combined_df = combined_df.astype('category')

#Define target columns
target_columns = [
    'P1ACTIVE', 'P4ACTIVE', 'P6ACTIVE', 'P7ACTIVE',
    'P1ATTENI', 'P4ATTENI', 'P6ATTENI', 'P7ATTENI',
    'P1BEHAVE', 'P4BEHAVE', 'P6BEHAVE', 'P7BEHAVE',
    'P1SOLVE', 'P4SOLVE', 'P6SOLVE', 'P7SOLVE', 'P6SAMEAG', 'P4SAMEAG'
]

#Check which target columns are present in the DataFrame
present_target_columns = [col for col in target_columns if col in combined_df.columns]

#Create a DataFrame to hold the target variables
target_df = combined_df[present_target_columns]

#Drop target columns from combined_df before one-hot encoding
combined_df = combined_df.drop(columns=present_target_columns)

#One-hot encode the remaining categorical variables, including handling NaNs
combined_df = pd.get_dummies(combined_df)

#Convert boolean columns to integers (0 and 1), if applicable
combined_df = combined_df.fillna(0).astype(int)

#Optionally, concatenate the target DataFrame back to the combined DataFrame
final_df = pd.concat([combined_df, target_df.reset_index(drop=True)], axis=1)
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()

#Fit and transform the data
columns_to_encode = target_df.columns

print("\nAFTER encoder")

final_df[columns_to_encode] = encoder.fit_transform(final_df[columns_to_encode])

#check the data types were changed 

for col in final_df.columns:
    print(final_df[col].value_counts())
```

```python
#Training set and Testing set
from sklearn.model_selection import train_test_split

#descriptive features 
X = final_df.drop(target_features, axis=1)

print(X.columns)
#target features
all_targets = final_df[target_df.columns]
X.head(2)
```

```python
#topics found in lecture 10
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score


y = all_targets['P7ATTENI']
print(y.isnull().sum())
#splitting data for train and test
mask = y.notna()
X = X[mask]
y = y[mask]
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.20, random_state = 42)

#performing preprocessing part
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)


#Applying PCA function on training and testing set of X component
pca = PCA(n_components=103)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
        
explained_variance_ratio = pca.explained_variance_ratio_
        

knn = KNeighborsClassifier(n_neighbors=30) #make KNN classifer instance
knn.fit(X_train, y_train) #train model on training set
y_pred = knn.predict(X_test) #test model 


#evaluating the accuracy 
knn_score = accuracy_score(y_test, y_pred)

knn = KNeighborsClassifier(n_neighbors=30, weights='distance')  # or try other parameters
scores = cross_val_score(knn, X_train, y_train, cv=5)
print("Avg cross-validation accuracy:", scores.mean())
print("Accuracy for train-test split: %0.3f"% knn_score)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 5))

# set x and y values
x = np.arange(1, len(explained_variance_ratio) + 1)
y = explained_variance_ratio

# plot
ax.plot(x, y, marker='o')
# set label and title
ax.set_xlabel('Principal Component')
ax.set_ylabel('Explained Variance Ratio')
ax.set_title('Explained Variance Ratio by Principal Component')

plt.xticks(np.arange(0, 150, 5))
plt.show()
```
